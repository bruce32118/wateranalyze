```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib
from dateutil.parser import parse 
import time
j = 0
while j < 100:
    testfile=urllib.URLopener()
    testfile.retrieve("http://waterpark.twd.gov.tw/www/opendata/wqb/wqb.asmx/GetQualityData","GetQualityData"+str(j))
    if j==0:
        f = open('GetQualityData'+str(j), 'r')
        print f
        soup = BeautifulSoup(f)

        a=soup.findAll(text=True) 
        h=u'\n'
        al=[]
        for i in a:
            if i != h:
                al.append(i)       
    
        count = 0
        name=[]
        cntu=[]
        cl=[]
        ph=[]
        longitude=[]
        latitude=[]
        for i in al:
            if count%9 == 3:
                name.append(i)
            if count%9 == 4:
                i=float(i)
                longitude.append(i)
            if count%9 == 5:
                i=float(i)
                latitude.append(i)
            if count%9 == 6:
                i=float(i)
                cntu.append(i)
            if count%9 == 7:
                i=float(i)
                cl.append(i) 
            if count%9 == 8:
                i=float(i)
                ph.append(i)
            if count == 1:
                watertime = i
            if count == 0:
                waterdate = i
            count = count+1            
        s = parse(waterdate+' '+watertime)
        place = pd.DataFrame(index=name)
        place['longitude']=longitude
        place['latitude']=latitude
        placeco=place.transpose()
        pd.DataFrame.to_excel(place,'place.xls')
        watercntu = pd.DataFrame(index=name)
        watercl = pd.DataFrame(index=name)
        waterph = pd.DataFrame(index=name)
        watercntu[s]=cntu
        watercl[s]=cl
        waterph[s]=ph
    else:
        f = open('GetQualityData'+str(j), 'r')
        print f
        soup = BeautifulSoup(f)

        a=soup.findAll(text=True) 
        h=u'\n'
        al=[]
        for i in a:
            if i != h:
                al.append(i)       
    
        count = 0
        name=[]
        cntu=[]
        cl=[]
        ph=[]
        for i in al:
            if count%9 == 3:
                name.append(i)
            if count%9 == 6:
                i=float(i)
                cntu.append(i)
            if count%9 == 7:
                i=float(i)
                cl.append(i) 
            if count%9 == 8:
                i=float(i)
                ph.append(i)
            if count == 1:
                watertime = i
            if count == 0:
                waterdate = i
            count = count+1            
        s = parse(waterdate+' '+watertime)
        watercntu[s]=cntu
        watercl[s]=cl
        waterph[s]=ph
    j=j+1
    time.sleep(900) 
watercntuco = watercntu.transpose()
waterclco = watercl.transpose()
waterphco = waterph.transpose()
pd.DataFrame.to_excel(watercntuco,'watercntucoo7.xls')
pd.DataFrame.to_excel(waterclco,'waterclcoo7.xls')
pd.DataFrame.to_excel(waterphco,'waterphcoo7.xls')
```
